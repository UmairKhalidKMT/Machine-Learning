```python
import numpy as np
import matplotlib.pyplot as plt

X=np.array([1.1,1.4,1.7,2.3,2.7,2.9])
y=np.array([2.2,2.5,2.7,3.6,4.9,5.1])

plt.scatter(X,y, color='red', label='Actual Data')
plt.xlabel('Sqaure Footage (1000 rfts)')
plt.ylabel('Price (100,000 $)')
plt.legend()
plt.show()
```

```python
def gradient_descent(X, y, m, b, learning_rate):
    m_gradient=0
    b_gradient=0
    n=len(X)
    for i in range(n):
        prediction=m*X[i]+b #using y=mx+b 
        b_gradient += -2 * (y[i] - prediction) #taking drivative w.r.t b
        m_gradient += -2 * X[i] * (y[i] - prediction) # taking drivative w.r.t m

    new_m -= (m_gradient / n) * learning_rate # updating parameters m
    new_b -= (b_gradient / n) * learning_rate # updating parameters b
    return new_m, new_b
```

```python
#now setting initial values and running gradient descent
m=0
b=0
learning_rate=0.01
epochs=1000 # number of iterations
for epoch in range(epochs):
    m, b = gradient_descent(X, y, m, b, learning_rate)
    if epoch % 100 == 0:
        print(f"Epoch {epoch}: m={m}, b={b}") # printing progress values every 100 epochs
        print(f"Prediction after {epoch} epochs: y={m*2.5 + b}") # predicting final price for 2500 sqft house
        
```

```python
plt.scatter(X, y, color='red', label='Actual Data')
plt.plot(X, m*X + b, color='blue', label='Fitted Line')
plt.xlabel('Square Footage (1000 rfts)')
plt.ylabel('Price (100,000 $)')
plt.legend()
plt.show()
```
